```
import random
import numpy as np
import matplotlib.pyplot as plt

m = 100   # number of points
x = 2 * np.random.rand(m, 1) 
yhat = 4 + 3 * x 
n = np.random.randn(m, 1) 
y = yhat + n

plt.plot(x,yhat, "b")
plt.plot(x,y, "k.")
plt.show()
```

```
# initial value
theta_0 = np.random.normal(0,1)
theta_1 = np.random.normal(0,1)

iterations = 1000
alpha = 0.1
j = []
theta0 = []
theta1 = []

for i in range(iterations):
    h = theta_0 + theta_1 * x
    
    temp_j = (1/(2*m))*sum((h-y)**2)
    j.append(temp_j)
    
    theta_0 = theta_0 - alpha*(1/m)*sum(h-y)
    theta0.append(theta_0)
    theta_1 = theta_1 - alpha*(1/m)*sum((h-y)*x)
    theta1.append(theta_1)
```

```
plt.plot(x,y, "k.")
plt.plot(x,h, "r")
plt.show()
```

```
plt.plot(j, "b")
plt.show()
```

```
plt.plot(theta0, "r")
plt.plot(theta1, "b")
plt.show()
```
